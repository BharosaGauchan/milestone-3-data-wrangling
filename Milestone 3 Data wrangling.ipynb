```python
import numpy as np
import pandas as pd
from urllib.request import urlopen
import json
import requests
```

```python
url =  'https://api.covidtracking.com/v1/states/daily.json'
response = requests.get(url)
print(response)
df = pd.DataFrame.from_dict(json.loads(response.text))
df.head()
```

```python
df.info()
```

```python
# selecting columns
df = df[['date', 'state','positiveTestsViral', 'negativeTestsViral','positiveCasesViral']] 
         #'hospitalizedCurrently' ,'hospitalizedCumulative',    
#'inIcuCurrently','inIcuCumulative','onVentilatorCurrently','onVentilatorCumulative']]   
df = df.rename(columns ={'state':'State'})
df
```

```python
df.isnull().sum()
```

```python
# Replacing NaN Values with 0
df = df.fillna(0)
# checking if there is any otherkind of null values is df_3
df.isnull().sum()

```

```python
# converting states to string dtype
#df['states']= df['states'].apply(str)
#df.info()
```

```python

df['date'] = df['date'].apply(lambda x: pd.to_datetime(str(x),format='%Y%m%d'))

df
```

```python
df2 = df.groupby([ 'State'])['positiveTestsViral', 'negativeTestsViral','positiveCasesViral'].sum().reset_index()
df2 = df2.astype({'positiveTestsViral':int, 'negativeTestsViral':int,'positiveCasesViral':int})
df2
```

```python
import sqlite3

#database setup
conn = sqlite3.connect('covid_df2.db')

#load df2 file to SQLite
# create table name
df2.to_sql('table2',conn, if_exists = 'replace')
# close the connection
conn.close()
```
